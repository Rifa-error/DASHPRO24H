```python
import numpy as np
import pandas as pd
np.random.seed(42)
jumlah_data = 1000
usia = np.random.randint(18,65, size=jumlah_data)
pendapatan = np.random.randint(2000000, 15000000, size = jumlah_data)
pengeluaran = np.random.randint(1000000, 10000000, size=jumlah_data)
skor_kredit = np.random.randint(300, 851, size = jumlah_data)
masa_kerja = np.random.randint(0, 30, size = jumlah_data)
jumlah_tanggungan = np.random.randint(0,6, size = jumlah_data)
jenis_pekerjaan = np.random.choice(['PNS','Swasta','Wirausaha','Profesional','Pejabat'], size = jumlah_data)
pendidikan = np.random.choice(['SMA','D3','S1','S2','S3'], size = jumlah_data)
status_pernikahan= np.random.choice(['Belum Menikah','Menikah','Cerai'], size = jumlah_data)

status_kredit=[]
for i in range(jumlah_data):
    if (skor_kredit[i] > 750) or\
       (skor_kredit[i] > 650 and pendapatan [i] > 2 * pengeluaran[i]) or \
      (masa_kerja[i] > 5 and pendapatan[i] > 1.5 * pengeluaran[i] and skor_kredit[i] > 500):
        status_kredit.append('Disetujui')
    else:
        status_kredit.append('Ditolak')

df = pd.DataFrame({
    'usia':usia,
    'pendapatan':pendapatan,
    'pengeluaran':pengeluaran,
    'skor_kredit':skor_kredit,
    'masa_kerja':masa_kerja,
    'jumlah_tanggungan':jumlah_tanggungan,
    'jenis_pekerjaan':jenis_pekerjaan,
    'pendidikan':pendidikan,
    'status_pernikahan':status_pernikahan,
    'status_kredit':status_kredit,
})

print(df)
print(df.info())
print(df.head())
df.to_csv("Data_Nasabah_Kredit.csv", index=False)
print("Data telah disimpan ke Data_Nasabah_Kredit.csv")
```
